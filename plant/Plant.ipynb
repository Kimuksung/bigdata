```
import pandas as pd
plant1_train = pd.read_csv("/gdrive/My Drive/Plain/plant1_train.csv")
plant2_train = pd.read_csv("/gdrive/My Drive/Plain/plant2_train.csv")
```

```
plant1_train = plant1_train.drop('Unnamed: 0' , axis=1)
cols = ['Date' , 'loc1_tem', 'loc1_hum', 'loc1_coil_temp','loc2_tem', 'loc2_hum', 'loc2_coil_temp' , 'loc3_tem', 'loc3_hum', 'loc3_coil_temp', 'out_tem', 'out_hum', 'loc1' , 'loc2' , 'loc3']
plant1_train.columns = cols
plant1_train_first = plant1_train[['Date','loc1_tem', 'out_tem' , 'loc1_coil_temp','loc1_hum' ,'out_hum','loc1']]
plant1_train_first["Hour"] = pd.to_datetime(plant1_train_first["Date"])
plant1_train_first["Hour"] = plant1_train_first["Hour"].apply(lambda x: x.hour)
tmp = []
for i in plant1_train_first["Hour"]:
  if 0<=i<6:
    tmp.append(1)
  elif 6<=i<12:
    tmp.append(2)
  elif 12<= i < 18:
    tmp.append(3)
  else:
    tmp.append(4)
plant1_train_first["Hour_label"] = tmp
plant1_train_first
```

```
plant1_train_second = plant1_train[['Date','loc2_tem', 'out_tem' , 'loc2_coil_temp','loc2_hum' ,'loc2']]
plant1_train_second["Hour"] = pd.to_datetime(plant1_train_second["Date"])
plant1_train_second["Hour"] = plant1_train_second["Hour"].apply(lambda x: x.hour)
tmp = []
for i in plant1_train_second["Hour"]:
  if 0<=i<6:
    tmp.append(1)
  elif 6<=i<12:
    tmp.append(2)
  elif 12<= i < 18:
    tmp.append(3)
  else:
    tmp.append(4)
plant1_train_second["Hour_label"] = tmp
plant1_train_second
```

```
plant1_train_third = plant1_train[['Date','loc3_tem', 'out_tem' , 'loc3_coil_temp','loc3_hum' ,'loc3']]
plant1_train_third["Hour"] = pd.to_datetime(plant1_train_third["Date"])
plant1_train_third["Hour"] = plant1_train_third["Hour"].apply(lambda x: x.hour)
tmp = []
for i in plant1_train_third["Hour"]:
  if 0<=i<6:
    tmp.append(1)
  elif 6<=i<12:
    tmp.append(2)
  elif 12<= i < 18:
    tmp.append(3)
  else:
    tmp.append(4)
plant1_train_third["Hour_label"] = tmp
plant1_train_third
```

24시간으로 전처리

```
tmp1 = plant1_train_first[plant1_train_first['Hour']==0]

tmp1['After24'] = tmp1['loc1'].shift(-1)
tmp1[tmp1['After24']==1]

tmp_loc1 = tmp1[tmp1['After24']==1]

for i in range(1,24):
  tmp1 = plant1_train_first[plant1_train_first['Hour']==i]

  tmp1['After24'] = tmp1['loc1'].shift(-1)
  tmp1 = tmp1[tmp1['After24']==1]  
  tmp_loc1 = pd.concat([tmp_loc1 , tmp1])

tmp_loc1
```

```
#tmp_loc1.drop(['Date', 'out_tem','loc1_coil_temp' , 'Hour_label'], axis=1).corr()
tmp_loc1 = tmp_loc1.drop(['Date', 'out_tem','loc1_coil_temp' , 'Hour_label'], axis=1)
tmp_loc1
```

```
tmp_loc1_not = plant1_train_first[plant1_train_first['loc1']==0].sample(500)
tmp_loc1_not = tmp_loc1_not.drop(['Date', 'out_tem','loc1_coil_temp' , 'Hour_label'], axis=1)
tmp_loc1_not['After24'] = 0.0
tmp_loc1_not
```

```
answer = pd.concat([tmp_loc1 , tmp_loc1_not] , axis= 0)
answer
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score  , confusion_matrix

#x_cols = ['loc1_tem', 'loc1_hum' , 'out_hum' , 'loc1' , 'Hour']
x_cols = ['loc1_tem', 'loc1_hum' , 'out_hum']
y_cols = ['After24']

lr = LogisticRegression()
model = lr.fit(X = answer[x_cols],y = answer[y_cols])
acc = model.score(answer[x_cols],answer[y_cols])
acc
```

```
from sklearn.svm import SVC
#params = [0.001 , 0.01, 0.1 , 1 ,10, 100] # 10^-3 ~ 10^2
#kernel = ["rbf" , "linear" , "poly"]
svc = SVC(gamma = 'auto')

model = svc.fit( answer[x_cols] , answer[y_cols])
y_pred = model.predict(answer[x_cols])
print(confusion_matrix(answer[y_cols], y_pred))

acc = model.score(answer[x_cols],answer[y_cols])
print(acc)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree.export import export_text 

dtc = DecisionTreeClassifier(criterion = "gini", max_depth=2)
x_cols = ['loc1_tem', 'loc1_hum', 'out_hum']
model = dtc.fit( answer[x_cols] , answer[y_cols] )
print(export_text(model , feature_names = x_cols))

y_pred = model.predict(answer[x_cols])
print(confusion_matrix(answer[y_cols], y_pred))

acc = model.score(answer[x_cols],answer[y_cols])
print(acc)
```
